Load data via pandas lib and take a glance on it

```python
import pandas as pd
import re
```

```python
data = pd.read_csv('titanic.csv', index_col='PassengerId')
data.head(10)
```

```python
males = data['Sex'].value_counts()[0]
females = data['Sex'].value_counts()[1]
print(males, females)
```

```python
survived_rate = data['Survived'].sum()/data['Survived'].shape[0]
died_rate = 1 - survived_rate
print(survived_rate, died_rate) 
```

```python
first_class_rate = data['Pclass'].value_counts()[1]/(data['Pclass'].value_counts().sum())
print(first_class_rate)
```

```python
mean_age = data['Age'].mean()
print(mean_age)
```

```python
median_age = data['Age'].median()
print(median_age)
```

```python
sibsp_parch_corr = data[['SibSp', 'Parch']].corr().as_matrix()[0, 1]
print(sibsp_parch_corr)
```

```python
for idx, name in enumerate(data['Name']):
    tmp_string = name
    tmp_string = re.split(',|\.| ', tmp_string)
#     print(tmp_string)
#     print(tmp_string[2]=='Mr')
#     print(tmp_string[1] == ('Miss.' or 'Mrs.'))
#     break
    if tmp_string[2] == 'Mr' or tmp_string[2] == 'Master':
        data.iloc[idx, 2] = 'Justamale'
    else:
        data.iloc[idx, 2] = tmp_string[4]
```
